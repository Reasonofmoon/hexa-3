## Step 1: 매장 정보 입력 (✏️)

```python
STORE_INFO = {
    'name': '✏️ [교육 기업명/매장명]',
    'type': '✏️ [오프라인매장|온라인쇼핑몰|O2O]',
    'platform': '✏️ [스마트스토어|쿠팡|자사몰|없음]',
    'employees': '✏️ [직원수]',
    'monthly_products': '✏️ [월 상품 등록 수]',
}
print('✅ 설정 완료:', STORE_INFO['name'])
```

## Step 2: AX 진단 (10개 문항, 각 1~5점)

| 점수 | 의미 |
|------|------|
| 1 | 전혀 안 함 |
| 3 | 부분적 수행 |
| 5 | 완전 자동화 |

```python
SCORES = {
    '상품 설명 자동화': 1,        # ✏️ 1~5
    'SEO 키워드 최적화': 1,       # ✏️ 1~5
    '고객 리뷰 모니터링': 1,      # ✏️ 1~5
    '재고 예측/관리': 1,          # ✏️ 1~5
    '가격 경쟁력 분석': 1,        # ✏️ 1~5
    '고객 세분화(CRM)': 1,       # ✏️ 1~5
    '이메일/SNS 자동화': 1,       # ✏️ 1~5
    '매출 리포트 자동화': 1,      # ✏️ 1~5
    '반품/CS 자동화': 1,          # ✏️ 1~5
    '채널별 통합 분석': 1,        # ✏️ 1~5
}
total = sum(SCORES.values())
level = '🟢 Starter(10~24)' if total<25 else ('🔵 Growing(25~39)' if total<40 else '🟣 AI-Ready(40~50)')
print(f'\n📊 총점: {total}/50점')
print(f'🎯 현재 레벨: {level}')
```

## Step 3: 시각화

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(list(SCORES.keys()),list(SCORES.values()),
        color=['#e74c3c' if v<=2 else '#f39c12' if v<=3 else '#2ecc71' for v in SCORES.values()])
ax.axvline(3,color='gray',linestyle='--',alpha=0.5,label='기준(3점)')
ax.set_xlim(0,5)
ax.set_title(f'{STORE_INFO["name"]} AX 진단 결과 — {level}')
ax.set_xlabel('점수 (1=미도입, 5=완전자동화)')
ax.legend()
plt.tight_layout()
plt.savefig('ax_diagnosis.png',dpi=150,bbox_inches='tight')
plt.show()
```

## Step 4: Gemini 맞춤 개선 추천

```python
weak=[k for k,v in SCORES.items() if v<=2]
prompt=f"""소매업/e커머스 AI 전환 컨설턴트로서 맞춤 전략을 작성하세요.
매장: {STORE_INFO['name']} ({STORE_INFO['type']})
총점: {total}/50점 — {level}
개선 필요 항목: {', '.join(weak) if weak else '전반적 고도화 필요'}
다음을 포함해 12주 로드맵을 작성하세요:
1. 즉시 시작 가능한 3가지 액션
2. 주요 도구 추천 (Gemini API 활용 포함)
3. 기대 효과 및 ROI
형식: 마크다운, 실용적이고 구체적으로"""
resp=model.generate_content(prompt)
print(resp.text)
```

## Step 5: 저장 & 다운로드

```python
with open('ax_report.md','w',encoding='utf-8') as f:
    f.write(f'# {STORE_INFO["name"]} AX 진단 리포트\n\n')
    f.write(f'**총점**: {total}/50점 | **레벨**: {level}\n\n')
    f.write('## 개선 추천\n\n')
    f.write(resp.text)
from google.colab import files
files.download('ax_report.md')
files.download('ax_diagnosis.png')
print('✅ 다운로드 완료! 다음은 W2: 상품 카피 자동화입니다.')
```
